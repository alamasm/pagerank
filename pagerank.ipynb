```python
import pandas as pd
import numpy as np
```

```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('data.xml')
root = tree.getroot()
```

```python
names = set()
for child in root[2][1][0]:
    print(child.attrib['source'])
    names.add(child.attrib['source'])
```

```python
names_ids = np.arange(len(names))
names_dict = dict(zip(names, names_ids))
```

```python
g = [[0 for i in range(len(names))] for j in range(len(names))]
```

```python
for child in root[2][1][0]:
    i = names_dict[child.attrib['source']]
    j = names_dict[child.attrib['target']]
    value = float(child.attrib['value'])
    g[i][j] = value
g
```

```python
page_rank = [1.0 for i in range(len(names))]
```

```python
def update_pagerank(g, page_rank, d):
    n = len(names)
    for i in range(n):
        pagerank_sum = 0
        for j in range(n):
            if (g[j][i] == 1.0):
                pagerank_sum += page_rank[j] / np.sum(g[j])
        page_rank[i] = d / n + (1 - d) * pagerank_sum
```

```python
for i in range(1000): update_pagerank(g, page_rank, 0.1)
```
